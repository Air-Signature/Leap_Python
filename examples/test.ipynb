```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
# Read data from CSV file
df = pd.read_csv('../Sign/5.csv')
```

```python
# Extract x, y, z coordinates
x2 = df['palm_x_coor']
y2 = df['palm_y_coor']
z2 = df['palm_z_coor']

# Create a 3D plot
fig2 = plt.figure(figsize=(8, 6))
ax2 = fig2.add_subplot(111, projection='3d')

# Plot the curve with improved marker style
# ax2.plot(x2, y2, z2, c='r', marker='o', markersize=5, linestyle='-')
# ax2.plot(x2, y2, z2, c='blue', marker='o', linestyle='-' ,ms = 4, mfc = 'deepskyblue', mec = 'deepskyblue')

ax2.plot(x2, y2, z2, ms = 4, mec = 'blue',linestyle='-', color = 'blue',
                                   picker=True , alpha=0.6)
ax2.plot(x2, y2, z2, marker = 'o', ms = 4, mec = 'b',linestyle='-', color = 'blue',
                                   picker=True , alpha=0.3)

# Set labels for each axis
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

# Add a title
ax2.set_title('3D Plot of Palm Coordinates')

# Add grid lines
ax2.grid(True)

# Show the 3D plot
plt.show()

# Create 2D projections
fig2, axs2 = plt.subplots(1, 3, figsize=(15, 5))



# X-Y projection
axs2[0].plot(x2, y2, c='blue',linestyle='-', 
                                   picker=True)
axs2[0].plot(x2, y2, c='r', marker = 'o', ms = 3, mfc = 'red', 
                                   picker=True , alpha=0.2)

axs2[0].plot(x2, y2, marker = 'o', ms = 4, mec = 'red',
                                   picker=True , alpha=0.5)
axs2[0].set_xlabel('X')
axs2[0].set_ylabel('Y')
axs2[0].set_title('X-Y Projection')
axs2[0].grid(True)

# X-Z projection
# X-Y projection
# X-Y projection
axs2[1].plot(x2, z2, c='blue',linestyle='-', 
                                   picker=True)
axs2[1].plot(x2, z2, c='r', marker = 'o', ms = 3, mfc = 'red', 
                                   picker=True , alpha=0.2)

axs2[1].plot(x2, z2, marker = 'o', ms = 4, mec = 'red',
                                   picker=True , alpha=0.5)
axs2[1].set_xlabel('X')
axs2[1].set_ylabel('Z')
axs2[1].set_title('X-Z Projection')
axs2[1].grid(True)

# Y-Z projection
axs2[2].plot(y2, z2, c='b', linestyle='-')
axs2[2].plot(y2, z2, c='b', marker='o', markersize=5, linestyle='-')

axs2[2].set_xlabel('Y')
axs2[2].set_ylabel('Z')
axs2[2].set_title('Y-Z Projection')
axs2[2].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the 2D projections
plt.show()
```

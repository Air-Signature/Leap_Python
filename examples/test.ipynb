```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
# Read data from CSV file
df = pd.read_csv('../Signatures/Object_Bird/1.csv')
```

```python


# Extract x, y, z coordinates
x = df['index_x_coor']
y = df['index_y_coor']
z = df['index_z_coor']


# Create a 3D plot plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y,z, c='r')


# Set labels for each axis
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the 3D plot
plt.show()

# Create 2D projections
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# X-Y projection
axs[0].plot(x, y, c='r', marker='o')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# X-Z projection
axs[1].plot(x, z, c='g', marker='o')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Z')

# Y-Z projection
axs[2].plot(y, z, c='b', marker='o')
axs[2].set_xlabel('Y')
axs[2].set_ylabel('Z')

# Show the 2D projections
plt.show()
```
